```python
import pandas as pd
```

```python
actuals_raw = pd.read_csv("fivethirtyeight_actuals.csv")
actuals = actuals_raw.set_index("team") * 100
pick_pcts_raw = pd.read_csv("yahoo_picks.csv")
pick_pcts = pick_pcts_raw.set_index("Team")
```

```python
class MarchCalculator():
    def __init__(self, actuals, pick_pcts):
        self.actuals = actuals
        self.pick_pcts = pick_pcts
    
    def round_value(self, team, round):
        win_prob = self.actuals.loc[team, str(round)]
        point_value = (2**(round - 2))
        pick_pct = self.pick_pcts.loc[team, str(round)]
        return (win_prob * (point_value / pick_pct))
    
    def cumulative_round_value(self, team, starting_round):
        rounds_to_eval = list(range(starting_round, 8))
        return sum([self.round_value(team, rd) for rd in rounds_to_eval])
    
    def evaluate_round(self, round):
        teams_to_eval = list(self.actuals.index)
        return pd.DataFrame(
            [[team, self.cumulative_round_value(team, round)] for team in teams_to_eval],
            columns=["team", "score"]
        ).set_index("team").sort_values("score", ascending=False)
```

```python
mc = MarchCalculator(actuals, pick_pcts)
mc.evaluate_round(2)
```
